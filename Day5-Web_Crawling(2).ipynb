- HTML: HyperText Markup Language

```html

<html>
    <head></head> : 문서에 표현되지는 않으나 문서 정의를 위해 필요한 데이터를 구조화
    <body>
        <h1>이것이 제목</h1> : h(heading text) 1단계 1~6
        <p>이것이 문단텍스트</p> : paragraph 문단텍스트
        <div></div> : html 태그를 구획화 할때
        <ul></ul> : unordered list 
        <ol></ol> : ordered list
        <main></main> : semantic tag, 문서의 메인컨텐츠를 정의할 때 사용
    </body> : 사용자가 볼 데이터(문서)를 구조화
</html> : 문서의 시작과 끝을 정의
```

```html
<section class="dynamic-text-list"></section> : 어떤 한 구역을 표현할 때 사용
```

````python
import requests
from bs4 import BeautifulSoup
import lxml

# 웹요청
tomato_uri = 'https://www.rottentomatoes.com/' # 요청 보낼 url
response = requests.get(tomato_uri) # requests로 요청 보내기
main_text = response.text # 응답의 html 텍스트를 할당

all_chart = [] # 모든 차트의 결과물을 담기위한 all_chart
# 분석시작
soup = BeautifulSoup(main_text, 'lxml') # BeutifulSoup 클래스 인스턴스 생성
sections = soup.find_all('section', attrs={'class':'dynamic-text-list'}) # class 값을 가진 section 태그 찾기: 4개
# 페이지에 존재하는 모든 차트에 대해
for i in range(4):
    section_box_office = sections[i] # 2번째의 box_office section 태그 확보
    lis = section_box_office.find_all('li') # 아래로 존재하는 모든 li(영화 한단위) 태그 확보
    box_office = [] # 결과값을 담을 그릇
    # for li in lis:
    for rank, li in enumerate(lis): # 순위 추가를 위한 index 생성 및 multi-assign으로 반복문 진행
        link = tomato_uri + li.find('a')['href'][1:] # href 속성 접근하기
        name = li.find('span').text # span 태그의 텍스트값 구하기
    
        try:
            #1 find_all (html)
            score = int(li.find_all('a')[1].text.strip(' \n%')) # 점수 구하기
            #2 select (CSS Selector)
            score = int(li.select('a:nth-child(2)')[0].text.strip(' \n%'))
        except ValueError as e:
            score = 0
        
        # print(link, name, score)
        box_office.append({
            'rank': rank+1,
            'link_url': link,
            'movie_name': name,
            'tomato_score': score,
        })
    all_chart.append(box_office)
````

### 순서
1. 페이지에 방문하여 웹브라우저 개발자모드(`F12` or `ctrl+shift+i` or `cmd+option+i`)로 요소의 특징 알아내기
2. 조사한 특징을 바탕으로 웹 요청 보내기(`requests`)
3. 문서텍스트에서 데이터 전체가 포함된 단위를 잘라낸 후, `0번째 단위`에 대해 샘플 작업수행(`beautifulsoup`)
4. 잘동작하면 나머지 단위들에 대해 반복적용(`for`)

````python
base_uri = 'https://www.rottentomatoes.com'
response = requests.get(base_uri + '/browse/movies_in_theaters/sort:popular')
response.status_code
````

````python
(2*2 + 26*4) > (26*2 + 2*4)
````

````python
# 요청과 분석 코드를 분리하여 불필요한 페이지 접속을 막을 수 있음
# 분석쪽 코드는 1초에 10만번 실행해도 문제되지 않음
soup = BeautifulSoup(response.text, 'lxml')

class_name = 'discovery-tiles__wrap'
movie_list = soup.find('div', attrs={'class':class_name})
# 1: find_all
# len(movie_list.find_all('div', attrs={'class':'js-tile-link'}))
# 2: select (using CSS Selector)
# len(movie_list.select('div.js-tile-link')) # . : class selector
movie_divs = movie_list.select('.js-tile-link') # 재생링크가 없는 것들은 div가 아니라 a태그여서 클래스 이름으로만 검색
# 0번째 요소에대해 테스트 후 전체적용
popular_movies = []

for movie_div in movie_divs:
    img_url = movie_div.find('rt-img')['src'] # 포스터 url
    movie_name = movie_div.find('rt-img')['alt'] # 영화제목
    if movie_div.name == 'div': # 2*2 + 26*4 // 26*2 + 2*4 # 그래서 div를 먼저 따짐(경제적)
        movie_url = base_uri + movie_div.find('a')['href'] # 영화 url
    else:
        movie_url = base_uri + movie_div['href'] # movie_div가 a태그라면 이렇게 접근
    opening_date = movie_div.find('span', attrs={'data-qa':'discovery-media-list-item-start-date'}).string.strip(' \n') # 개봉일
    # print(img_url, movie_name, movie_url, opening_date)
    popular_movies.append({
        'movie_name': movie_name,
        'poster_url': img_url,
        'movie_url': movie_url,
        'opening_date': opening_date,
    })
````

## Selenium

- requests+BeautifulSoup: 정적인 페이지(Static Page) -> http response(응답)으로 모든 데이터가 준비되는 경우
- 동적페이지(Dynamic Page)
    - template(HTML): 구조
    - data(API): 구조에 끼워넣을 데이터
    - 페이지가 이미 로드된 상태에 추가로 데이터가 날아오는 경우
    - 사용자의 액션으로 데이터가 로드되는 경우

- https://data.kleague.com/: 메인페이지
- https://data.kleague.com/: 데이터포털
- https://data.kleague.com/: 공식기록
- https://data.kleague.com/: 역대전적 -> requests로 수집할 수 없는 사이트

## 스크래핑을 하고자 할 때 flowchart
- 내가 원하는 데이터를 공식 API로 받을 수 있는가?
    - Yes, API key를 발급받아 사용
    - No, 해당 사이트 개발자도구-Network에서 데이터 request 흔적이 발견되었는가?
        - Yes, requests 로 url request를 보내 데이터 수집
        - No, 사이트 내 페이지 이동시 url이 변경되는가? and 샘플로 수집을 해봤을 때 데이터가 표현되는가?
            - Yes, requests + BeautifulSoup
            - No, Selenium(동적페이지 or iframe or 사용자 액션(키보드, 마우스 입력)이 필요 ..)

### requests -- selenium

- requests
    - 내가 `어떤 페이지(주소)`에서 어떤 요소에 접근해야 하는지를 조사
