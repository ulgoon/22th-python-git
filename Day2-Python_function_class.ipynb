````python
## Monty Hall

from random import shuffle, randint, choice

stay = 0
switch = 0

# n회 반복시행
times = 10000
for _ in range(times):
    doors = [0, 0, 1,]
    shuffle(doors)
    # print(doors)
    
    # guest = doors[randint(0,2)]
    # guest = choice(doors)
    guest = doors.pop(randint(0,2))
    # print(guest)
    # print(doors)
    
    # 염소 한마리 골라낸 뒤, (doors.remove(0))
    doors.remove(0)
    if guest == 1:
        # 지켰을때 (stay += 1, guest==1)
        stay += 1
    elif doors[0] == 1:
        # 바꿨을때 (switch += 1, doors[0]==1)
        switch += 1
    

# 결과 출력
print(f'Stay: {stay}/{times} ({stay/times*100}%)')
print(f'Switch: {switch}/{times} ({switch/times*100}%)')
````

## Leap Year

- input: 0~9999 사이의 정수
- output: (사용자의 입력년도, 윤년 평년 여부)

- 4로 나누어 떨어지면 윤년
- 100으로 나누어 떨어지면 평년
- 400으로 나누어 떨어지면 윤년

````python
year = 2002
(year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
````

````python
def is_leap(year: int) -> tuple:
    """
    입력한 연도의 윤년 여부를 판단합니다.
    """
    # 다중조건문 ver
    # if year % 400 == 0:
    #     is_leap = True
    # elif year % 100 == 0:
    #     is_leap = False
    # elif year % 4 == 0:
    #     is_leap = True
    # else:
    #     is_leap = False
    # 논리연산자 ver
    is_leap = (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)
    return year, is_leap
````

````python
# Research 를 통해 Memoization 구현해보기
# hint: mem = {0:0, 1:1}

def fibo_mem():
    # memoization(어떤 시퀀스의 실행결과를 저장하는 메모리)
    mem = {0:0, 1:1}
    def sub_func(n):
        if not n in mem:
            mem[n] = sub_func(n-1) + sub_func(n-2)
        return mem[n]
    return sub_func
f_m = fibo_mem()
f_m(7)
````

````python
def outer_scope():
    def inner_local():
        msg = 'Kernel'

    def inner_nonlocal():
        nonlocal msg
        msg = 'Academy'

    def inner_global():
        global msg
        msg = 'KernelAcademy'
    msg = 'Seoul'
    print(msg, id(msg))
    inner_local()
    print('local', msg, id(msg))
    inner_nonlocal()
    print('nonlocal', msg, id(msg))
    inner_global()
    print('global', msg, id(msg))
````
