````python
# boolean indexing
# 2022년의 평균연령이 45.5세 이상인 모든 구 선택
# df # df
# df['2022'] # column
# df['2022'] >= 45.5 # 비교연산

df[df['2022'] >= 45.5] # 비교연산 결과가 True인 구만 남음
````

````python
# 먼저해보기
# 2019년의 평균연령이 45.0세 미만인 모든 구
df
df['2019']
df['2019'] < 45.0
df[df['2019'] < 45.0]
# 2020년의 평균연령이 44.5세 초과인 모든 구
df
df['2020']
df['2020'] > 44.5
df[df['2020'] > 44.5]
````

````python
# 2022년의 평균연령이 45.5세 이상 (or) 2019년의 평균연령이 45.0세 미만인 모든 구
# df['2022'] >= 45.5
# df['2019'] < 45.0
# df[df['2022'] >= 45.5 or df['2019'] < 45.0]
df[(df['2022'] >= 45.5) | (df['2019'] < 45.0)] # bitwise operator: |
df[(df['2022'] >= 45.5) & (df['2019'] < 45.0)] # bitwise operator: &
````

````python
## groupby

data = {
"Team": ["Sales 1", "Sales 2", "Sales 1", "Sales 2", "Sales 1", "Sales 2", "Sales 1", "Sales 2"],
"Quarter": ["Q1", "Q1", "Q2", "Q2", "Q3", "Q3", "Q4", "Q4"],
"Sales Volume": [1500, 1200, 1650, 1300, 1800, 1400, 2000, 1600],
"Revenue": [750, 600, 825, 650, 900, 700, 1000, 800],
"Profit": [180, 140, 200, 160, 220, 180, 250, 210],
}
df = pd.DataFrame.from_dict(data)
````

````python
df.assign(
    Team_Avg_Sales=df.groupby('Team')['Sales Volume'].transform('mean'),
    Team_Sum_Sales=df.groupby('Team')['Sales Volume'].transform('sum'),
    Team_Min_Sales=df.groupby('Team')['Sales Volume'].transform('min'),
    Team_Max_Sales=df.groupby('Team')['Sales Volume'].transform('max'),
)
````

````python
#filter

# df.groupby('Team')
df.groupby('Team').filter(lambda x: x['Profit'].sum() > 700)

# Team을 기준으로 grouping 한 뒤, Revenue의 최대값이 850 미만인 팀을 고르시오
df.groupby('Team').filter(lambda x: x['Revenue'].max() < 850)
# Team을 기준으로 grouping 한 뒤, Profit의 합계가 700 미만인 팀을 고르시오
df.groupby('Team').filter(lambda x: x['Profit'].sum() < 700)
# Team을 기준으로 grouping 한 뒤, Sales Volume의 평균이 1500 이상인 팀을 고르시오
df.groupby('Team').filter(lambda x: x['Sales Volume'].mean() >= 1500)
````

````python
employees = {
"EmployeeID": [101,102,103,104],
"Name": ["김민준","이서연","박지훈","최예나"],
"Team": ["Sales 1","Sales 2","Sales 1","Marketing"]
}
performance = {
"EmployeeID": [ 101, 102, 103, 105 ],
"Score": [ 95, 88, 92, 80 ],
"Salary (Million KRW)": [ 60, 55, 62, 50 ],
"Team": [ "Sales 1", "Sales 2", "Sales 1", "HR" ]
}
````

````python
df1 = pd.DataFrame.from_dict(employees)
df2 = pd.DataFrame.from_dict(performance)
````

````python
df1
````

````python
df2
````

````python
set(df1.columns) & set(df2.columns)
````

````python
set(df1.columns) | set(df2.columns)
````

````python
pd.merge(df1,df2, on='EmployeeID', how='inner')
````

````python
pd.merge(df1,df2, on='EmployeeID', how='left')
````

````python
pd.merge(df1,df2, on='EmployeeID', how='outer')
````

````python
pd.merge(df1, df2, on='Team', how='inner', suffixes=('_emp', '_perf'))
# suffixes: 어떤 df의 데이터였는지 남기기 위한 구분자
````

````python
df_result = pd.merge(df1, df2, on='Team', how='inner', suffixes=('_emp', '_perf'))
````

````python
# save
# df_result.to_csv('./data/inner_join_employees.csv')
df_result.to_csv('./data/inner_join_employees.csv', index=False) # range index를 제거하기위한
df_result.to_csv('./data/inner_join_employees_utf-8.csv', index=False, encoding='utf-8') # 웹공유(universal)
df_result.to_csv('./data/inner_join_employees_cp949.csv', index=False, encoding='cp949') # MS Office
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
````

````python
titanic = sns.load_dataset('titanic')
type(titanic)
````

````python
titanic.head()
````

````python
plt.figure(figsize=(6,4))
sns.histplot(data=titanic, x='fare', bins=30, kde=True)
plt.title('운임 분포')
plt.show()
````

````python
plt.figure(figsize=(6,4))
sns.kdeplot(data=titanic, x='age', fill=True)
plt.title('나이 분포(밀도)')
plt.grid(True, linestyle=':', alpha=0.3)
plt.show()
````

````python
plt.figure(figsize=(7,5))

sns.scatterplot(data=titanic, x='age', y='fare', hue='survived')
plt.title('상관관계(나이-요금)')
plt.show()
````

````python
corr_matrix = titanic[['age', 'fare', 'pclass', 'sibsp']].corr()

plt.figure(figsize=(7,6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='icefire')
plt.title('상관관계(요금,나이,클래스)')
# xG, xA, G, A, 평균활동량
plt.show()
````

````python
plt.figure(figsize=(7,5))
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('분포(나이-등급)')
plt.show()
````

````python
plt.figure(figsize=(7,5))

sns.violinplot(data=titanic, x='sex',y='age')
plt.title('성별-나이 분포')
plt.show()
````
