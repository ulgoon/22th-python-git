## Elevator class 구현하기

어떤 건물의 엘리베이터가 1대 있습니다. 엘리베이터에 대해 클래스를 정의하고, 인스턴스를 생성하세요

1. 공간(space): `[](instance 변수)`
2. 상승(go_up()): method
3. 하강(go_down()): method
4. 현재상태(status): 점검중, 만원, 운행가능(instance 변수)
5. 현재위치(floor): instance 변수

```python
class Elevator:
    """
    엘리베이터를 구현하는 클래스
    status: a(available), m(maintenance mode), f(full)

    #TODO: 운행가능층(min_floor,max_floor), 중량, 엘리베이터의 현재위치 세분화
    여러 층수 누르기(눌러진 버튼들을 저장할 변수 필요(stops))
    응급버튼 -> 점검모드, 버저를 울리게 함
    중량에 대한 처리
    
    여러 인스턴스(동기화된 엘리베이터가 존재한다.) -> 여러 엘베가 있을 때 나에게 가장 가까운 or 나를향해 달려오는 엘베 멈추게 하기
    """
    def __init__(self, space, floor=1, status='a'):
        self.space = space
        self.floor = floor
        self.status = status

    def go_up(self, target_floor):
        """
        엘리베이터의 상승
        # TODO: 몇명이 타는지를 parameter로 받아서 space, 중량 증가
        max_floor 범위 안에 있는지
        현재위치보다 낮은 층수를 눌렀을때에 대한 처리
        """
        print(f'현재 {self.floor}층입니다.')
        self.floor = target_floor
        print(f'{self.floor}층으로 이동하였습니다.')
    def go_down(self, target_floor):
        """
        엘리베이터의 하강
        """
        print(f'현재 {self.floor}층입니다.')
        self.floor = target_floor
        print(f'{self.floor}층으로 이동하였습니다.')
```

```python
class Wallet:
    def __init__(self, account):
        self.account = account

    balance = 0
    name = 'Your Wallet'

    @classmethod
    def set_default(cls, amount):
        while amount < 1:
            print('Error: Default amount should be more than 1. Try again!!!!')
            amount = int(input("Enter Default Value: "))
        cls.balance = amount
        print(f'Your balance is now {cls.balance}')

    def add_to_account(self, amount):
        self.account += amount
        print(f'Total balance: {self.account + Wallet.balance}')

    @staticmethod
    def see_static_name():
        print(Wallet.name)

    def get_class_name(cls):
        print(cls.name)


class MyWallet(Wallet):
    name = 'My wallet'
```

```python
try:
    num1, num2 = [int(s) for s in input('정수 두개를 띄어쓰기로 구분하여 입력하세요: ').split(' ')]
    print(num1 / num2 + '3')
# except:
#     print('손님 여기서 이러시면 안됩니다')
except TypeError as e:
    print(f'연산할 수 없는 타입입니다. {e}')
except ValueError as e:
    print(f'정수가 아닌 값을 입력하시면 안됩니다. {e}')
except ZeroDivisionError as e:
    print(f'0을 입력하시면 안됩니다: {e}')

print('Program closed')
```

```python
# try구문을 성공적으로 수행했을 경우 else 블록을 실행
try:
    num1, num2 = [int(s) for s in input('정수 두개를 띄어쓰기로 구분하여 입력하세요: ').split(' ')]
    result = num1 / num2
# except:
#     print('손님 여기서 이러시면 안됩니다')
except TypeError as e:
    print(f'연산할 수 없는 타입입니다. {e}')
except ValueError as e:
    print(f'정수가 아닌 값을 입력하시면 안됩니다. {e}')
except ZeroDivisionError as e:
    print(f'0을 입력하시면 안됩니다: {e}')
else:
    print(f'실행결과: {result}')
print('Program closed')
```

```python
# finally: try든 except든 해당블록 처리를 마치고 방문하는 블록
# session
# 파일을 연다 -> 작업을 한다 -> 파일을 저장하고 닫는다.
try:
    num1, num2 = [int(s) for s in input('정수 두개를 띄어쓰기로 구분하여 입력하세요: ').split(' ')]
    result = num1 / num2
# except:
#     print('손님 여기서 이러시면 안됩니다')
except TypeError as e:
    print(f'연산할 수 없는 타입입니다. {e}')
except ValueError as e:
    print(f'정수가 아닌 값을 입력하시면 안됩니다. {e}')
except ZeroDivisionError as e:
    print(f'0을 입력하시면 안됩니다: {e}')
else:
    print(f'실행결과: {result}')
finally:
    print('어쨋거나 try 구문 실행을 완료했습니다')
print('Program closed')
```
