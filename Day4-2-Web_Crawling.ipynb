## Requests

```python
import requests
```

```python
# requests.get('https://www.google.com/') # 웹 정보 요청
response = requests.get('https://www.google.com/') # 웹 정보 요청으로 반환되는 Response 객체를 변수에 할당
```

```python
response.status_code
# Status code 체크를 항상 해서 어떤 문제가 발생했을 때 요청으로 인한 문제가 아님을 확실히 소거
# 100, 200(OK), 300(301-회원가입시 url 변동으로인한 OK),// 400(client error), 500(server error)
# 404 Not found -
# 403 Forbidden
# 503 Service Unavailable
```

```python
type(response.text)
```
