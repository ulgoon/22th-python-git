```python
## while: 조건반복문
```

```python
status = True

# while문을 사용할때는 status를 False로 만들 조건을 확실히 정하고 작성및수행을 해야 합니다.
while status: # status가 True인 동안
    greeting = input("안녕?") # 기계가 인사하는데
    if greeting == '안녕': # 인사를 받아주면
        status = False # 종료
    else:
        print('넌 정말 나쁜아이구나') #다른말 하면 계속 인사
```
